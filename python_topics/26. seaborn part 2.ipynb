### Categorical Scatter plot:

- Normally `Scatterplot` is used in case of numerical data but it can also be used over categorical data.
- But in this case one column should be of numerical and another of categorical type.

```python
# To check which column is numerical and which is categorical

tips.info()
```

```python
# Doing same with iris dataset

iris.info()
```

```python
# Normal scatter plot
# Here both columns are numerical

sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

### strip plot()

- This is an **Axes Level Function**
- It take one categorical and one numerical column and then create a scatterplot between them.

```python
# Here "day" in x axis is of categorical type
# And "total_bill" in y axis is of numerical type

sns.stripplot(data=tips, x='day', y='total_bill')
plt.show()
```

**Notes:**

- Here we can see the data points a bit distributed instead of they being in a single line.
- It is because `seaborn` adds some noise so the data points don't get into a single line.
- To get them into a single line we need to add a parameter `jitter` with value `False`.

```python
# Making the same above plot with data points in single line

sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
plt.show()
```

### catplot()

- This is a **Figure Level Function**.

```python
# Now doing same using catplot()
# Here "kind" is "strip"

sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
plt.show()
```

```python
# Now using jitter
# Here we can manipulate the distribution of the data points using values in jitter

sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=0.2)
plt.show()
```

```python
# Using hue

sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=0.2, hue='sex')
plt.show()
```

### swarmplot()

- This is **Axes Level Function**
- Although it is almost same as the `stripplot()` but here it represents the data points in such a way that they represent the distribution of the data points in the dataset.
- So here the distribution of data is represented in a more better way.

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
# Doing same using catplot()
# Here "kind" is "swarm"

sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
plt.show()
```

```python
# using hue

sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

### Categorical Distribution Plots

### Boxplot

- A boxplot is a standardized way of displaying the distribution of data based on a five number summary (**minimum**, **first quartile [Q1]**, **median**, **third quartile [Q3]** and **maximum**). 
- It can tell you about your outliers and what their values are. 
- Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

- Although `boxplot` shows the distribution of one column but here using `seaborn` we can plot it with another categorical column.

```python
# Here we are ploting "total_bill" with the "day" column using boxplot() 

sns.boxplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
# Doing same using catplot()
# Here "kind" is "box"

sns.catplot(data=tips, x='day', y='total_bill', kind='box')
plt.show()
```

```python
# hue

sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
# single boxplot -> numerical col
# It shows the distribution of one single column as here "total_bill"

sns.boxplot(data=tips, y='total_bill')
plt.show()
```

```python
# We can also do it in horizontally by changing the axis

sns.boxplot(data=tips, x='total_bill')
plt.show()
```

```python
# Doing same using catplot()
# Here "kind" is "violin"

sns.catplot(data=tips, x='day', y='total_bill', kind='violin')
plt.show()
```

```python
# hue

sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex')
plt.show()
```

### Categorical Estimate Plot

### barplot()

- Using this we can calculate some aggregate value of a numerical column with another categorical column.
- When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars. These are the balck bars at the top. 
- These error bars are shown when the value of the x axis gets repeated multiple times in the data. This basically calculates the Confidence Interval.
- To remove the error bars we need to pass a parameter `ci` as `None`.

```python
import numpy as np
```

```python
# Here we want to find the avg total bill of males and females

sns.barplot(data=tips, x='sex', y='total_bill')
plt.show()
```

```python
# Here we see the same without the black bars
# We pass a parameter "ci" as "None"

sns.barplot(data=tips, x='sex', y='total_bill', ci=None)
plt.show()
```

```python
# hue
# Now doing seperation on the basis of smoker

sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker')
plt.show()
```

```python
# estimator
# By default it's value is "mean"
# Here we are calculating the minimum value instead of average

sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', ci=None, estimator=np.min)
plt.show()
```

```python
# Doing same using catplot()

sns.catplot(data=tips, x='sex', y='total_bill', hue='smoker', kind='bar', ci=None, estimator=np.min)
plt.show()
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

### regplot()

- This is **Axes Level Function**.
- Here `hue` parameter is not available.

```python
# Getting regression line between tip and total bill

sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```

### lmplot()

- This is **Figure Level Function**.
- Here `hue` is available.

```python
sns.lmplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# hue
# Here we will get two different regression lines for each gender

sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

- Here the dotted line represents the correct result and the other data points show their error in predicting the regression line for each data point.

### FacetGrid

- Upto here all the 3 **Figure Level Functions** i.e. `relplot()`, `displot()` and `catplot()` has the Facet column options.
- There is another way to create Facet columns, and that is known as **FacetGrid**.
- It gives a bit more flexibility i.e. more customizable.

```python
# Creating facet grid on the basis over "day" and "time"

sns.catplot(data=tips, x='sex', y='total_bill', kind='bar', col='day', row='time')
plt.show()
```

```python
# Now creating same graph using FactGrid
# Here first we create a FacetGrid object using the row and column
# Then we use "map()" over that object and pass the plotname and column names we want in x axis and y axis
# "add_legend()" to add the lebels

g = sns.FacetGrid(data=tips, col='day', row='time', hue="smoker")
g.map(sns.barplot,'sex','total_bill')
g.add_legend()
plt.show()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)
- The `pairplot()` gives relation (`scatterplot`) pairwise between numerical columns of the dataset.
- Here when it show relation between two different columns it plots a `scatterplot` and when it compare the same column with itself (the diagonal graphs) it shows a `histogram`.
- When we use `hue` parameter we get `KDEplot` in place of `histograms`.
- Pairplot is a special case of PairGrid.

### PairGrid
- Here the Pairplot internally gets created due to PairGrid.
- The way to create it is same as FacetGrid.
- Here the flexibility is we can create different graphs for diagonal and non diagonal items using `map_diag()` and `map_offdiag()` functions.

```python
# Creating pairplot on the basis of species

sns.pairplot(iris)
plt.show()
```

```python
# Now if we use "hue" 

sns.pairplot(iris, hue='species')
plt.show()
```

```python
# Now creating same graph using PairGrid
# Here it will create all scatterplot

g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
plt.show()
```

```python
# map_diag() & map_offdiag()
# Here we are creating same graph but here we will create histplot for all diagonal 
# i.e. when comparing the column with itself
# For this we will use "map_diag()" with the plot name we want
# And for non diagonal i.e. comparing with other columns we create scatterplot using "map_offdiag()"

g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
plt.show()
```

```python
# Now with hue 
# Here we are changing the plots

g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
plt.show()
```

- We can also create different plots for above and below the diagonal plots.
- For this we use `map_upper()` and `map_lower()`.
- Also we can restrict on which columns we want to work with using the `vars` parameter.

```python
# map_diag -> map_upper -> map_lower
# Here diagonals will have histograms
# The upper plots of the diagonals will have 2d KDEplots
# The lower plots of diagonals will be scatterplots

g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
plt.show()
```

```python
# vars
# Here we will work with only 2 columns instead of 4

g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
plt.show()
```

### JointGrid

### JointGrid Vs Jointplot

- The `jointplot()` creates `scatterplot` between two numerical columns along with the `histogram` of each of the two columns we passed. This is by default.
- We can also change the graphs with the `kind` parameter.
- We can do the same using `JointGrid()`.
- Here again the rules are same as above.

```python
# Creating a jointplot between total bill and tip

sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# kind
# Here we will get 2d histogram and 1d histogram.

sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist')
plt.show()
```

```python
# We can also get 2d KDEplot and 1d KDEplot

sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
plt.show()
```

```python
# Now we will get regression and kde+histogram

sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
plt.show()
```

```python
# Geting hexplot along with histogram

sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
plt.show()
```

```python
# hue

sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist', hue='sex')
plt.show()
```

```python
# Now creating using JointGrid()

g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.histplot)
plt.show()
```

```python
# Now customizing

g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.kdeplot, sns.boxplot)
plt.show()
```
