### Seaborn Roadmap

**Types of Functions**

- **Figure Level Function**
- **Axis Level Function**
- We can create plots using both kind of functions but they have their own advantages/disadvantages. 
- The hierarchy is:
  - Figure
  - Multiple Axis/Graphs in the Figure
  - Multiple objects (e.g. labels, line, markers, xticks/yticks) in each Graph

**Main Classification**

- Relational Plot: It is used when we need to plot the relation between multiple quantities.
  - Scatter plot (Relation between two quantities)
  - Line plot (Mainly for Time Series)
- Distribution Plot: It is used when we need to plot graph on a single quantity to see it's distribution.
  - Histogram
  - KDE plot
  - Rug plot
- Categorical Plot: Mainly used on Categorical data
  - Bar plot
  - Count plot
  - Box plot
  - Violin plot
  - Swam plot
- Regression Plot: To plot the regression predictions.
  - Reg plot
- Matrix Plot: This is used over grid of data.
  - Heatmap
  - Clustermap
- Multiplots: 
  - Joint plot
  - Pair plot

### 1. Relational Plot

- To see the statistical relation between 2 or more variables/quantities/columns.
- Mainly used for `Bivariate Analysis`.

Plots under this section

- scatterplot
- lineplot

```python
# Loading the dataset

tips = sns.load_dataset('tips')
tips
```

### scatter plot()
- This is an **Axes level function**

```python
# "hue" will differentiate the points in the plot based on the values of column name passed as value
# "style" to add another layer of information

plt.figure(figsize=(12, 9))
sns.scatterplot(data=tips, x='total_bill', y='tip', 
                hue='sex', style='time', size='size')

plt.show()
```

### relplot()

- This is a **Figure Level Function**

**Notes:**
- Now creating the same `Scatterplot`sing a Figure level function named `relplot()`
- Here we need to pass the plot type to a parameter named `kind`. The rest code is same.
- Generally the Figure level function creates the graph in a bit squarish shape by default.

```python
# relplot()
# Here "kind" is "scatter"

sns.relplot(data=tips, x='total_bill', y='tip', 
            kind='scatter', hue='sex', style='time', size='size')

plt.show()
```

**Notes:**

- According to the library recommendation one should use more Figure Level Functions as their hierarchy is higher. 
- That is by using one Figure Level Function many Axes Level Functions can be replecated.
- When using Figure Level Function the `legend` stays out of the graph area.
- Whereas when using Axis Level Function the `legend` will be inside the graph area.
- **Remember:** If it is an Axes Level Function then we can reshape the graph plot in the same way as we reshape the figure size in `matplotlib`, but for Figure Level Function it is not possible using matplotlib.
- To do this we need  to use the parameters `height` and `aspect` parameters.

### line plot()
- This is an **Axes Level Function**

```python
# Loading the dataset
# Here px.data.gapminder() is to load the data from plotly library

gap = px.data.gapminder()
gap
```

```python
# Creating a dataset for India only

temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# Here we are checking the life expectancy in India from 1962 to 2007

sns.lineplot(data=temp_df, x='year', y='lifeExp')
plt.show()
```

```python
# Doing the same using relplot() which is a figure level function
# Her "kind" is line

sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
plt.show()
```

```python
# Getting data of "India", "Brazil" and "Germany"

temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
# Using "hue", "style", "size"

sns.relplot(data=temp_df, x='year', y='lifeExp', 
            kind='line', hue='country', style='continent', 
            size='continent')

plt.show()
```

```python
# Doing same using lineplot()

sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
plt.show()
```

### Facet plot()

- Using this we can create multiple plots based on a particular categorical column.
- To do this instead of using the `hue` parameter we need to use the `col` parameter.
- Also we can create the same using `row` parameter.
- For `col` the graphs will be side by side and for `row` they are up down.
- This only works with **Figure Level Functions**
- So this works with `relplot()`
- But it will not work with `scatterplot()` and `lineplot()`.

```python
# Creating a scatterplot using relplot()
# Here on the basis of gender that many scatterplots will get created
# Here the graphs will be side by side

sns.relplot(data=tips, x='total_bill', y='tip', 
            kind='scatter', col='sex')
plt.show()
```

```python
# Doing same using row
# Here the graphs will be up down

sns.relplot(data=tips, x='total_bill', y='tip', 
            kind='scatter', row='sex')
plt.show()
```

```python
# We can also do based on both
# That is in column we want to see whether he is a smoker
# On row we want to see his gender

sns.relplot(data=tips, x='total_bill', y='tip', 
            kind='scatter', col='smoker', row='sex')
plt.show()
```

```python
# Now creating day wise and gender wise

sns.relplot(data=tips, x='total_bill', y='tip', 
            kind='scatter', col='sex', row='day')
plt.show()
```

```python
# Doing same using kind as line

sns.relplot(data=tips, x='total_bill', y='tip', 
            kind='line', col='sex', row='day')
plt.show()
```

```python
# If we try to create facet plot using scatterplot()

try:
    sns.scatterplot(data=tips, x='total_bill', y='tip', col='sex', row='day')
except Exception as err:
    print(err)
```

```python
# Creating scatterplot between life expectancy and gdp per capita

sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter')

plt.show()
```

```python
# Now if we want scatterplot for each year differently

sns.relplot(data=gap, x='lifeExp', y='gdpPercap', 
            kind='scatter', col='year')

plt.show()
```

```python
# col_wrap
# But it is hard to read 
# So to overcome this we can wrap the columns using "col_wrap" parameter
# Here we want 3 columns in each row

sns.relplot(data=gap, x='lifeExp', y='gdpPercap', 
            kind='scatter', col='year', col_wrap=3)

plt.show()
```

### histplot()

- This is **Axes Level Function**.

```python
tips.head()
```

```python
# plotting univariate histogram of the "total_bill"

sns.histplot(data=tips, x='total_bill')
plt.show()
```

### displot()

- This is **Figure Level Function**.

```python
# Now creating the same histogram using "displot()"
# Here "kind" is "hist"

sns.displot(data=tips, x='total_bill', kind='hist')
plt.show()
```

```python
# bins parameter
# Here creating 5 bins

sns.displot(data=tips, x='total_bill', kind='hist', bins=5)
plt.show()
```

**Notes:**

- Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
- Discrete bins are automatically set for categorical variables.
- This is technically not a histogram, we can call it as a `countplot`.

```python
# Plotting histogram on categorical data column "day"
# Here we want to check how many customers came for each day

sns.displot(data=tips, x='day', kind='hist') 
plt.show()
```

```python
# hue parameter
# Here we can see which customer paid more tip based on gender

sns.displot(data=tips, x='tip', kind='hist', hue='sex')
plt.show()
```

```python
# element -> step
# To clear the image we need to use the parameter "element" with value as "step"
# Here instead of creating solid image it creates just the outlines

sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
plt.show()
```

```python
# Loading dataset

titanic = sns.load_dataset('titanic')
titanic
```

```python
# We want to create histogram of age genderwise

sns.displot(data=titanic, x='age', kind='hist', hue='sex')
plt.show()
```

```python
# Now using the "element" 

sns.displot(data=titanic, x='age', kind='hist', element='step', hue='sex')
plt.show()
```

```python
# Now creating facet plots
# Remember we cannot create facet plot using histplot() as it is an Axis Level Function

sns.displot(data=titanic, x='age', kind='hist', element='step', col='sex')
plt.show()
```

### kdeplot()

- This is **Kernal Density Estimation**
- This is **Axes Level Function**.
- Here rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate.
- So basically instead of creating bins it creates a curve to show the distribution of the data.
- Here the `y-axis` represents the probability  of the density in the data.
- It is more useful than `histplot` as it shows continuous curve.

```python
# Creating distribtion of total bill
# So here there is a 5% chance that most total bills are around 10

sns.kdeplot(data=tips, x='total_bill')
plt.show()
```

**Doing same using `displot()`**

```python
# Here "kind" is "kde"

sns.displot(data=tips, x='total_bill', kind='kde')
plt.show()
```

```python
# hue -> To see genderwise distribution

sns.displot(data=tips,x='total_bill', kind='kde', hue='sex')
plt.show()
```

```python
# fill -> to fillup with color
# The value should be "True"

sns.displot(data=tips,x='total_bill', kind='kde', hue='sex', fill=True)
plt.show()
```

**To manipulate the size of the Figure Level Function**

```python
# height -> Height (in inches) of each facet. 
# So here the image becomes of 6 inches

sns.displot(data=tips,x='total_bill', kind='kde', hue='sex', fill=True, height=6)
plt.show()
```

```python
# aspect -> it will create an aspect according to the "height" parameter
# Here the width will be double of "height"

sns.displot(data=tips,x='total_bill', kind='kde', hue='sex', fill=True, height=10, aspect=2)
plt.show()
```

### rugplot()

- This is **Axes Level Function**.
- Not mostly used.
- Plot marginal distributions by drawing ticks along the x and y axes.
- This function is intended to complement other plots by showing the location of individual observations in an unobtrusive (to be seen differently) way.
- That is it is use for support for another plot to better visualization.

```python
# Here we create a kdeplot() of total bill
# Then we do a "rugplot()" to see the same information in different form

sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
plt.show()
```

### Bivariate histogram

- Here we can create histogram of two columns.
- A bivariate histogram bins the data within rectangles that tile the plot.
- Then shows the count of observations within each rectangle with the fill color.

```python
# Here creating histogram of tips and total bill

sns.histplot(data=tips, x='total_bill', y='tip')
plt.show()
```

**Notes**
- Here two histograms got created horizontally for `tip` and vertically for `total_bill` and wherever they intersect we can call them as **Grid**.
- The darker the color of the **Grid** represents the more number of data points fall in that area.

```python
# Doing same using "displot()"

sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
plt.show()
```

### Bivariate Kdeplot

- A Bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian.

```python
# Doing same for kdeplot

sns.kdeplot(data=tips, x='total_bill', y='tip')
plt.show()
```

### 2. Matrix Plot

- Heatmap
- Clustermap

### heatmap()

- This is **Axes Level Function**
- It plot rectangular data as a color-encoded matrix.

```python
# Here we want to plot the life expectancy of each country for each year
# Creating a pivot to make it wide format
# So here the shape of grid is "number of countries X number of years"

temp_df = gap.pivot(index='country', columns='year', values='lifeExp')
temp_df
```

```python
# So we are representing the grid as a heatmap
# As it is an Axes Level Function so we are reshaping the figure size

plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
plt.show()
```

```python
# Creating same graph for only European countries

temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
temp_df
```

```python
# Now creating same heatmap

plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
plt.show()
```

```python
# annot -> to see the values in numbers
# To see this we need to pass value as "True"

plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True)
plt.show()
```

```python
# linewidth -> to create a gap between each boxes to improve readability

plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True, linewidth=0.5)
plt.show()
```

```python
# cmap -> to change the color scheme
# Here we are using "viridis"

plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True, linewidth=0.5, cmap='viridis')
plt.show()
```

### clustermap()

- It is very similar to `heatmap`.
- It plot a matrix dataset as a hierarchically-clustered heatmap.
- It makes similar columns as a group (clustering) and dissimilar columns different.
- This function requires `scipy` to be available.
- Here we need to pass the column names on which we want to perform the clusters.

```python
# importing dataset

iris = px.data.iris()
iris
```

```python
# Here we want to find out similarity of these 4 columns in the dataset.
# Also we want to check the similarity between the 150 flowers.
# Now creating a clustermap to do the above

sns.clustermap(iris.iloc[:, [0,1,2,3]])
plt.show()
```

**Notes:**

- These are called **Dendograms**.
- Here it changes the order of the columns based on their similarity.
- Then doing same with the rows also.
- Internally here a ML works named as **Agglomerative Hierarchical Clustering**.
