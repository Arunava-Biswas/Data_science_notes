### Colored Scatterplots

```python
# Importing the dataset

iris = pd.read_csv('datasets/datasets-session-24/iris.csv')
iris.sample(5)
```

```python
# Making the column "Species" as numerical

iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 
                                           'Iris-versicolor':1,
                                           'Iris-virginica':2})
iris.sample(5)
```

```python
# Making Scatterplot

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

### `cmap`, `alpha` and `colorbar()`

- To make each color different need to pass the species to parameter `c` which represents colour.
- Remember the `Species` needed to be in numeric format. 
- The `cmap` to provide a colormap.
- The `alpha` to give opacity of the graph.
- The `colorbar()` is to represent which color represents which value.

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],
            c=iris['Species'], cmap='jet', alpha=0.7)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

### Plot size

- To change the size of the graph we can use `figure()`.
- In this we need to pass the sizes as tuple inside a parameter `figsize` like `figsize(width, height)`.

```python
# Changing the size

plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],
            c=iris['Species'], cmap='jet', alpha=0.7)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

### Annotations

```python
# Importing another dataset

batters = pd.read_csv('datasets/datasets-session-24/batter.csv')
batters
```

```python
# Making a sample of random 25 batsmen from top 100 batsmen from the dataset

sample_df = batters.head(100).sample(25, random_state=5)
sample_df
```

```python
# Creating a normal scatter plot

plt.scatter(sample_df['avg'], sample_df['strike_rate'])
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Average vs Strike Rate")
plt.show()
```

**Notes:**

- Here in the graph we cannot differentiate which batsmen is which dot.
- To solve this problem we can use annotations. Using which we can place text on the graph.
- To give label we need to use a function named `text()` which takes two parameters (the coordinates) and the label.
- We can also by using `fontdict` parameter we can change the size, color etc of the point.

```python
# Here creating 2 arrays
x = [1,2,3,4]
y = [5,6,7,8]

# Now creating a scatterplot between x and y
plt.scatter(x,y)
plt.text(1,5,'Point 1')  # here 1 and 5 act as parameter for point 1
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4', fontdict={'size':12, 'color':'brown'})
plt.show()
```

```python
# Now applying the same logic we will create the annotation for each batsmen
# Here "sample_df['batter'].values[i]" for the batsman's name

plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'], sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])

plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Average vs Strike Rate")
plt.show()
```

```python
# Now suppose we want to decide the size of the dot of the batsman's name based on the runs scored by him
# So we will use the parameter "s" and passing the runs as value

plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])

plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Average vs Strike Rate")
plt.show()
```

### Horizontal and Vertical lines

- To put horizontal and vertical lines on the graph.
- We use `axhline()` for horizontal line and `axvline()` for vertical line.
- We can also give different colors to the lines using parameter `color`.

```python
# Want to fetch only those batsmen whose Strikerate is between 130 and 140 and avg is > 30

plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

plt.axhline(130, color='red')
plt.axhline(140, color='blue')
plt.axvline(30, color='green')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])

plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.title("Average vs Strike Rate")
plt.show()
```

### Subplots

- To plot multiple plots side by side at one go.

```python
# A diff way to plot graphs

batters.head()
```

```python
# Normal way of making plot

plt.figure(figsize=(15,6))
plt.scatter(batters['avg'], batters['strike_rate'])
plt.title('Avg vs SR')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
# Another way
# The "plt.subplots()" returns two objects
# Here "fig" is for the figure that is the graph
# Here "ax" is for the axes that is the 'x' and 'y' axis
# Then we call the scatter using the axes object
# Here to add the title we need to do "set_title()"
# Also for labeling 'x' and 'y' axis we need to use "set_xlabel()" and "set_ylabel()"
# To change the figure size we need to specify the "figsize()" in the "subplots()"
# Need to pass "fig.show()" instead of "plt.show()"

fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax.set_title('Avg vs SR')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

**Creating multiple plots**

```python
# The "nrows" is for number of rows
# The "ncols" is for number of columns
fig, ax = plt.subplots(nrows=2, ncols=1)
```

```python
# The "fig" is for the graph

fig
```

```python
# The axis will have two objects

ax
```

```python
# Here we create two graphs side by side
# The "ax[0]" is for the 1st graph and "ax[1]" for second graph
# The "sharex" or "sharey" is to share the x axis or y axis between all the graphs. 
# For this pass value as "True"
# As here both the graphs are sharing same x axis  so we use "sharex" and pass value as "True"

fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10,6))

ax[0].scatter(batters['avg'], batters['strike_rate'], color='red')
ax[1].scatter(batters['avg'], batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')

fig.show()
```

```python
# Here we are creating 4 graphs first 2 as scatter and last 2 as histograms
# Here we get a 2d array so to fetch the 1st graph we need to pass 2 numbers

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

ax[0,0].scatter(batters['avg'], batters['strike_rate'])
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

fig.show()
```

```python
# Another way to do the same
# Here we create a figure object using "plt.figure()"
# Then we can create different subplots using the "fig.add_subplot()"
# Here (2,2,1) means 2 rows and 2 columns and this is the 1st graph

fig = plt.figure()

# 1st graph
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'], batters['strike_rate'], color='red')

# 2nd graph
ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

# 3rd graph
ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])

# 4th graph
ax4 = fig.add_subplot(2,2,4)
ax4.scatter(batters['avg'], batters['runs'])

fig.show()
```

### 3D Surface Plots

- Mainly used in Machine Learning to visualize loss functions.

```python
# Creating a graph for x^2 + y^2
# For this creating 2 arrays using linspace

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
# 100 points of x

x
```

```python
# 100 points of y

y
```

```python
# Now creating a grid using "meshgrid()"
# It will rteturn 2 axes "xx" and "yy" each with a shape of (100,100)

xx, yy = np.meshgrid(x,y)

print("Shape of xx: ", xx.shape)
print("Shape of yy: ", yy.shape)
```

```python
# Now calculating "z" using the meshgrid

z = xx**2 + yy**2
print("Shape of z: ", z.shape)
```

```python
# Now creating the surface plot using "plot_surface()"
# Here in colorbar we need to pass the variable where we store the graph

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)

ax.set_xlabel('xx')
ax.set_ylabel('yy')
ax.set_zlabel('z')

fig.show()
```

```python
# Another graph of sin and cos
# Here z = sin(x) + cos(y)

z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)

ax.set_xlabel('xx')
ax.set_ylabel('yy')
ax.set_zlabel('z')

fig.show()
```

```python
# Now z = sin(x) + log(y)

z = np.sin(xx) + np.log(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)

ax.set_xlabel('xx')
ax.set_ylabel('yy')
ax.set_zlabel('z')

fig.show()
```

### Contour Plots

- This is a special graph that represents the 3d graph into 2d.
- Mainly used in Machine Learning to visualize loss functions.

```python
# Making the z = x^2 + y^2 graph into contour
# For this we use "contour()"
# Here there will be no z level as it is a 2d plot

z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx, yy, z, cmap='viridis')
fig.colorbar(p)

ax.set_xlabel('xx')
ax.set_ylabel('yy')

fig.show()
```

```python
# Another way is to use "contourf()"
# This is a filled graph

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(p)

ax.set_xlabel('xx')
ax.set_ylabel('yy')

fig.show()
```

```python
# Doing same for sin(x) + sin(y)

z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(p)

ax.set_xlabel('xx')
ax.set_ylabel('yy')

fig.show()
```

### Heatmap

- Here we transform a grid into graph where in grid it shows how values are distributed.

```python
# Importing another dataset

delivery = pd.read_csv('datasets/datasets-session-24/IPL_Ball_by_Ball_2008_2022.csv')
delivery
```

```python
# Checking number of balls in one over
# As there can be extra balls due to noballs and wides

delivery['ballnumber'].unique()
```

```python
# Creating a dataset for legitimate balls where 6 was hit

temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
temp_df
```

**We want to create a grid where we want to print in each ball of each over how many 6s had hit**

```python
# Now creating a pivot table to find number of 6s for each ball in each over

grid = temp_df.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count')
grid
```

```python
# Now plotting it
# To create a heatmap we need to use "imshow()" 
# where we need to pass the variable name of the grid we want to transform to heatmap
# here list(range(1,21)) is used so we can get from 1st to 20th over as index in y axis
# list(range(1,7)) to get ball number from 1 to 6 in x axis

plt.figure(figsize=(20,10))

plt.imshow(grid)

plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()

plt.show()
```
