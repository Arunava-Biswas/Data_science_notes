# Violin Plot

- It is almost same as the `boxplot` and `whisker plot` of matplotlib.
- It shows the distribution of quantitive data accross several levels of one or more categorical variables so the distributions can be compared.
- To do this we need to use `violinplot()` along with `x`, `y` and `data` paramteres.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Now importing the dataset

```python
df = sns.load_dataset("tips")
df
```

```python
# Now do the plotting
# Here we want to find out on which day how many people paid how much bill

plt.title("Per day bill by customers", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

**Notes**

- Here the fat part of the plot shows where the concentration of data points is highest. Like here the most number of tips are paid between `10` and `20` and that also most number of tips are paid on `Thursday`, as it has the most fatter design.

**Parameter `hue`**

- To differentiate the data points.

```python
# To know whether the number of lunch or dinner is high?

plt.title("Per day bill by customers", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df, hue="time")
plt.show()
```

**Parameter `linewidth`**

- To thicken the border of the graph.

```python
plt.title("Per day bill by customers", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df, linewidth=3)
plt.show()
```

**Parameter `palette`**

- To customize the colors.

```python
plt.title("Per day bill by customers", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df, linewidth=3, palette="Dark2")
plt.show()
```

**Parameter `order`**

- To change the order of the graph.
- In `order` we need to provide the values as list as we areusing more than one values.

```python
# How much amount we got at Lunch vs Dinner?

plt.title("Dinner vs Lunch bills", fontsize=20)
sns.violinplot(x="time", y="total_bill", data=df, linewidth=3, palette="Dark2", order=["Dinner", "Lunch"])
plt.show()
```

**Parameter `saturation`**

- To lighten the graph.
- The value range is from 0 to 1.

```python
plt.title("Per day bill by customers", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df, linewidth=3, palette="Dark2", saturation=0.6)
plt.show()
```

**Parameter `color`**

- To change the color of the bars.

```python
plt.title("Per day bill by customers", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df, linewidth=3, color="y")
plt.show()
```

**Parameter `split` and `scale`**

- To make more seperation use `split`.
- In `scale` we can use `count`, `area`, `width` as values. 

```python
# To differentiate between Male and Female on the basis of lunch and dinner.

plt.title("Dinner vs Lunch bills genderwise", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df, hue="sex", split=True)
plt.show()
```

```python
plt.title("Dinner vs Lunch bills genderwise", fontsize=20)
sns.violinplot(x="day", y="total_bill", data=df, hue="sex", split=True, scale="count")
plt.show()
```

- To make the plots Horizontal just change the data points of the axises. That is interchange the data points of `x-axis` and `y-axis`.

```python
plt.title("Dinner vs Lunch bills genderwise horizontal", fontsize=20)
sns.violinplot(x="total_bill", y="day", data=df, hue="sex")
plt.show()
```

**Parameter `inner`**

- When we use `quart` as a value then the part inside the vilon get distributed in quartiles.
- We can also use values `box`, `quartile`, `point`, `stick`, `None`. By default the value is `box`.

```python
plt.title("Dinner vs Lunch bills", fontsize=20)
sns.violinplot(x="time", y="total_bill", data=df, order=["Dinner", "Lunch"], inner="quart")
plt.show()
```

### To create a single violin plot

- For this we only need to take a numeric feature.
- Also we will need only the `x` axis here.
- We also need a series instead of a dataframe to be passed in the `x` axis.
- To make it vertically we need to change the `x` axis with `y` axis.

```python
plt.title("Single violin plot for Total Bills", fontsize=20)
sns.violinplot(x=df["total_bill"])
plt.show()
```

```python
plt.title("Horizontal Single violin plot for Total Bills", fontsize=20)
sns.violinplot(y=df["total_bill"])
plt.show()
```
