# Histogram Plot

- Using seaborn if we want to plot a histogram we need to use the `displot()` i.e. the **Distribution Plot**.
- Remember to make a plot using seaborn function we need to pass dataframe in it.
- There is another `distplot()` it will create a distribution plot along with the **`Kernal Density Estimator (KDE)`**. But this plot is deprecated.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Now importing dataset

```python
dataset1 = sns.load_dataset("penguins")
dataset1
```

```python
# Now doing the plot using 'displot()'

sns.displot(dataset1['flipper_length_mm'])
plt.show()
```

**Parameter `bins`**

- To fix the range of the data.

```python
# Here creating bins from 170 to 240

sns.displot(dataset1['flipper_length_mm'], 
            bins=[170, 180, 190, 200, 210, 220, 230, 240])
plt.show()
```

**Parameter `kde`**

- To create a line graph along with the bar graph.
- To see the `kde` the `bins` parameter is not necessarry.

```python
sns.displot(dataset1['flipper_length_mm'], 
            bins=[170, 180, 190, 200, 210, 220, 230, 240], 
            kde=True)
plt.show()
```

**Parameter `rug`**

- To see the data points of the `kde` on the x-axis.

```python
sns.displot(dataset1['flipper_length_mm'], 
            bins=[170, 180, 190, 200, 210, 220, 230, 240], 
            kde=True, rug=True)
plt.show()
```

**Parameter `color`**

- To change the color of the bars.

```python
sns.displot(dataset1['flipper_length_mm'], 
            bins=[170, 180, 190, 200, 210, 220, 230, 240], 
            kde=True, rug=True, color="r")
plt.show()
```

**Parameter `log_scale`**

- To change the data points in log.

```python
sns.displot(dataset1['flipper_length_mm'], 
            bins=[170, 180, 190, 200, 210, 220, 230, 240], 
            kde=True, rug=True, color="g", log_scale=True)
plt.show()
```

```python
# removing the 'bins' to show the log scale

sns.displot(dataset1['flipper_length_mm'], 
            kde=True, rug=True, color="g", log_scale=True)
plt.show()
```

```python
# Now doing the plot using 'distplot()'

sns.distplot(dataset1['flipper_length_mm'])
plt.show()
```
