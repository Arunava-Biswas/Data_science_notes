# Count Plot

- To do count plot we use `countplot()` with arguement `data` and `x`.
- Here x-axis will provide the number of category in that feature and y-axis will provide the count of each category in that feature.
- Count plot also provides a Bar graph but the differences are:
    - `Countplot` plots the count of the number of records by category.
    - `Barplot` plots a value or metric for each category.
    - By default `Barplot` plots the mean of a variable by category.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Now importing dataset

```python
dataset1 = sns.load_dataset("tips")
dataset1
```

```python
# Now plotting

plt.title("Count of Male and Female", fontsize=20)
sns.countplot(x="sex", data=dataset1)
plt.show()
```

### If we used `barplot()`

```python
plt.title("Mean of Male and Female using Bar plot", fontsize=20)
sns.barplot(x="sex", y="size", data=dataset1)
plt.show()
```

**Parameter `hue`**

- Here we can seperate the data on the basis of another feature.
- Like here we want to know number of smokers genderwise.

```python
plt.title("Count of smokers both Male and Female", fontsize=20)
sns.countplot(x="sex", data=dataset1, hue="smoker")
plt.show()
```

### To change the bar to horizontally

- Change the axis from `x` to `y`.

```python
plt.title("Horizontal Count of smokers both Male and Female", fontsize=20)
sns.countplot(y="sex", data=dataset1, hue="smoker")
plt.show()
```

**Parameter `palette`**

- To change the colours of the bar.

```python
plt.title("Count of smokers both Male and Female", fontsize=20)
sns.countplot(x="sex", data=dataset1, hue="smoker", palette="bwr")
plt.show()
```

**Parameter `edgecolor` and `linewidth`**

- To make differences between the bars.

```python
plt.title("Count of smokers both Male and Female", fontsize=20)
sns.countplot(x="sex", data=dataset1, hue="smoker", palette="bwr", linewidth=3, edgecolor="black")
plt.show()
```

### To get the values along with the count plot:

- Pass the countplot in a variable as here `ax`.
- Iterate the returned axes of the countplot and show the count values at the top of the bars. For this use the `annotate()`.

```python
plt.title("Count of smokers both Male and Female", fontsize=20)
ax = sns.countplot(x="sex", data=dataset1, hue="smoker", palette="bwr")

for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

plt.show()
```
