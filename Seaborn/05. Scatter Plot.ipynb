# Scatter Plot

- Here to create scatter plot we need to use `scatterplot()` with parameters `x`, `y` to create the axis and `data` for the dataframe.
- Scatter plots are mainly use to show the distribution of the data points.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Now importing dataset

```python
dataset1 = sns.load_dataset("penguins")
dataset1
```

```python
# now creating the plot

plt.title("bill_length vs bill_depth", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=dataset1)
plt.show()
```

**Parameter `hue`**

- To see the differentiation in the data points.

```python
plt.title("bill_length vs bill_depth genderwise", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=dataset1, hue="sex")
plt.show()
```

**Parameter `style`**

- To change the style of the graph.
- We need to pass the name of the feature through which we want to style the graph.
- Like here we want to change the styles of the gender.

```python
# Here the male are represented as 'o' and female as 'X'

plt.title("bill_length vs bill_depth genderwise", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", 
                data=dataset1, hue="sex", style="sex")
plt.show()
```

**Parameter `size`**

- Changing the size of the data points.
- Here also we can pass the name of the feature through which we want the size to be calculated, as here it is `sex` feature.

```python
plt.title("bill_length vs bill_depth genderwise", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", 
                data=dataset1, hue="sex", style="sex", size="sex")
plt.show()
```

**Parameter `sizes`**

- To customize the size according to our needs.
- Remember to use this first we need to use the `size` parameter.

```python
# Here the size of male datapoints are 40 and female are 100

plt.title("bill_length vs bill_depth genderwise", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=dataset1, 
                hue="sex", style="sex", size="sex", sizes=(100, 40))
plt.show()
```

**Parameter `palette`**

- To change the colors.
- To know the values use the logic we used in previous plots.

```python
plt.title("bill_length vs bill_depth genderwise", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=dataset1, 
                hue="sex", style="sex", size="sex", 
                sizes=(100, 40), palette="gist_ncar")
plt.show()
```

**Parameter `alpha`**

- To blur the plot.

```python
plt.title("bill_length vs bill_depth genderwise", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=dataset1, 
                hue="sex", style="sex", size="sex", sizes=(100, 40), 
                palette="gist_ncar", alpha=0.7)
plt.show()
```

**Parameter `markers`**

- To change the markers showing the data points, As here `male` are represented by `o` and `female` by `x`.
- To change this first we need to create a dictionary and put the markers as value.
- Now we want to show `male` as `*` and `female` as `o`.

```python
# creating the marker
m = {'Male': "*", 'Female': "o"}

plt.title("bill_length vs bill_depth genderwise", fontsize=20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=dataset1, 
                hue="sex", style="sex", size="sex", sizes=(100, 40), 
                palette="gist_ncar", alpha=0.7, markers=m)
plt.show()
```
