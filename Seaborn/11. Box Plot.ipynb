# Box Plot

- Here we use `boxplot()` with `x`, `y` and `data` parameters.
- We can make box plot also from matplotlib. But is seaborn we can use it in a more better way.
- In box plot we categorize the data into quartiles and then we can find the mean of that data using the box.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
df = sns.load_dataset("tips")
df
```

```python
# 1st we need to create a grid for box plot
# Here we are using "whitegrid" to create the grid

sns.set(style="whitegrid")
```

```python
# Now plotting the data

plt.title("Day vs Total Bill", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

**Notes**

- The dots at the head works as the outliers.

**Parameter `hue`**

- To sepearate the datapoints according to need

```python
plt.title("Day vs Total Bill based on meal", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="time")
plt.show()
```

```python
plt.title("Day vs Total Bill based on gender", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex")
plt.show()
```

**Parameter `color`**

- To change the color of the boxes.

```python
plt.title("Day vs Total Bill based on gender", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", color="g")
plt.show()
```

**Parameter `order`**

- To change the order of the boxes.

```python
plt.title("Day vs Total Bill based on gender", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", color="g", order=["Fri", "Sun", "Thur", "Sat"])
plt.show()
```

**Parameter `showmeans`**

- To show the means of the boxes.
- This will show the mean of each boxes.
- To differentiate the colors let's change the color to red.

```python
plt.title("Day vs Total Bill based on gender", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", 
            color="r", order=["Fri", "Sun", "Thur", "Sat"], showmeans=True)
plt.show()
```

**Parameter `meanprops`**

- To customize the properties like marker, color etc related to mean.

```python
plt.title("Day vs Total Bill based on gender", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", 
            color="r", order=["Fri", "Sun", "Thur", "Sat"], 
            showmeans=True, meanprops={'marker': "+", 'markeredgecolor': "b"})
plt.show()
```

**Parameter `linewidth`**

- To increase the border widths.

```python
plt.title("Day vs Total Bill based on gender", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", 
            color="r", order=["Fri", "Sun", "Thur", "Sat"], 
            showmeans=True, meanprops={'marker': "+", 'markeredgecolor': "b"},
           linewidth=3)
plt.show()
```

**Parameter `palette`**

- To customize the colors of the boxes.
- Here we can remove the `color` parameter as here we can get more customized colors.

```python
plt.title("Day vs Total Bill based on gender", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", 
            order=["Fri", "Sun", "Thur", "Sat"], 
            showmeans=True, meanprops={'marker': "+", 'markeredgecolor': "b"},
           linewidth=3, palette="plasma")
plt.show()
```

**Parameter `orient`**

- To change the direction of the boxes. By default it is vertical.
- Now to make the boxes horizontal we need both the data points to be `numeric` type.
- To make it horizontal place the value as `h`.

```python
plt.title("Day vs Total Bill based on gender in default orientation", fontsize=20)
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", 
            order=["Fri", "Sun", "Thur", "Sat"], 
            showmeans=True, meanprops={'marker': "+", 'markeredgecolor': "b"},
           linewidth=3, palette="plasma", orient="v")
plt.show()
```

```python
# Interchanging the axis values without using "orient"

plt.title("Day vs Total Bill based on gender by changing the axis values", fontsize=20)
sns.boxplot(x="total_bill", y="day", data=df, hue="sex", 
            order=["Fri", "Sun", "Thur", "Sat"], 
            showmeans=True, meanprops={'marker': "+", 'markeredgecolor': "b"},
           linewidth=3, palette="plasma")
plt.show()
```

```python
# Now with using "orient"

plt.title("Day vs Total Bill using orient", fontsize=20)
sns.boxplot(data=df, 
            showmeans=True, meanprops={'marker': "+", 'markeredgecolor': "b"},
           linewidth=3, palette="plasma", orient="h")
plt.show()
```

### Making a single boxplot

```python
plt.title("Single box plot of total bill", fontsize=20)
sns.boxplot(x=df["total_bill"])
plt.show()
```

```python
# To change the direction just change the axis

plt.title("Vertical Single box plot of total bill", fontsize=20)
sns.boxplot(y=df["total_bill"])
plt.show()
```
