# Pair Plot

- This plot is used to make all the comparisons in the dataset in one go.
- To create this plot we use `pairplot()`.
- It creates the plots between numerical values only by default.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
df = sns.load_dataset("tips")
df
```

```python
# Now making the plots

sns.pairplot(df)
plt.show()
```

**Parameter `hue`**

- To make differentiation in the plot.

```python
sns.pairplot(df, hue="sex")
plt.show()
```

**Parameter `vars`**

- To set the features we want in our plot.
- As here we only create plots between two features on the basis of gender.

```python
sns.pairplot(df, vars=["total_bill", "tip"], hue="sex")
plt.show()
```

**Parameter `hue_order`**

- To set the order of the `hue` parameter.
- Now at first it will show female and then male.

```python
sns.pairplot(df, vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"])
plt.show()
```

**Parameter `palette`**

- To set the color of the graph.

```python
sns.pairplot(df, vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"], palette="BuGn")
plt.show()
```

**Parameter `x_vars` and `y_vars`**

- To change the variable axis.
- To make it work we need to clear the `vars` parameter that we used in earlier case.
- Same to do with `y_vars` also

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], x_vars=["total_bill", "tip"])
plt.show()
```

**Parameter `kind`**

- To change the kind of the graphs.
- Here we can pass any one of the values `scatter`, `kde`, `hist`, `reg`.

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], kind="reg")
plt.show()
```

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], kind="kde")
plt.show()
```

**Parameter `diag_kind`**

- To set different kinds fixed in the diagonal plots.
- In the previos case all the plots are of `kde` kind.
- As here we now can see kde plot along with histograms in the diagonal plots.

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], kind="kde", diag_kind="hist")
plt.show()
```

**Parameter `markers`**

- To change the markers of the data points.

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], markers=["*", "<"])
plt.show()
```
