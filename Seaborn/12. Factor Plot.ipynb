# Factor Plot

- Here we use `factorplot()` with `x`, `y` and `data` parameters.
- With the help of this data we can only create a point plot.
- Here there is a parameter called `kind`, with the help of which we can create different categorical plots using this `factorplot`.
- So we can create all the plots using this plot.
- This plot is deprecated, instead we need to use `catplot()`.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
df = sns.load_dataset("tips")
df
```

```python
# Now plotting
# Remember to put the plot title i.e. 'plt.title()' after creating the plot

sns.factorplot(x="size", y="tip", data=df)
plt.title("Normal Point Plot of Size vs Tip", fontsize=20)
plt.show()
```

**Parameter `hue`**

- To differentiate the data points.

```python
sns.factorplot(x="size", y="tip", data=df, hue="sex")
plt.title("Normal Point Plot of Size vs Tip genderwise", fontsize=20)
plt.show()
```

**Parameter `kind`**

- To create different plots.

```python
sns.factorplot(x="size", y="tip", data=df, hue="sex", kind="bar")
plt.title("Bar graph of Size vs Tip genderwise", fontsize=20)
plt.show()
```

```python
sns.factorplot(x="size", y="tip", data=df, hue="sex", kind="box")
plt.title("Box plot of Size vs Tip genderwise", fontsize=20)
plt.show()
```
