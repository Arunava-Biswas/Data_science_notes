# Facet Grid

- This is used for **Multiplotting**. That is when we need to plot 2 or more than 2 plots inside one plot.
- To do this we use `FacetGrid()` and pass the dataframe to it.
- Then we need to use the `map()` and inside that we can pass the plot name we want to plot.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
df = sns.load_dataset("tips")
df
```

```python
# Creating different plots in the same plot

sns.boxplot(y="total_bill", data=df)
plt.title("Vertical Box plot of Total Bill", fontsize=20)
plt.show()
```

#### Using Facetgrid

- Remember we need to first create an object of the `FacetGrid()`.
- To plot the data we will use the `map()`.
- In map we need to pass the plot name we want.

```python
# Here we will see single plot
fg = sns.FacetGrid(df)

# Now plotting the dataset
fg.map(plt.scatter, "total_bill", "tip")
plt.title("Scatter plot of Total Bill vs tips", fontsize=20)
plt.show()
```

**Parameter `col`**

- To create multiplots we need to pass this parameter to the object.
- Here we can see different graphs on the basis of gender.

```python
# Now we will see multiple plots genderwise
fg = sns.FacetGrid(df, col="sex")


# Now plotting the dataset
fg.map(plt.scatter, "total_bill", "tip")

plt.show()
```

**Parameter `hue`**

- To create more differentiation in the plot.

```python
# Now we will see multiple plots genderwise and daywise
fg = sns.FacetGrid(df, col="sex", hue="day")


# Now plotting the dataset
fg.map(plt.scatter, "total_bill", "tip")

plt.show()
```

**Parameter `add_legend()`**

- We need to add this function in map to see the different names.

```python
# Now we will see multiple plots genderwise and daywise
fg = sns.FacetGrid(df, col="sex", hue="day")


# Now plotting the dataset
fg.map(plt.scatter, "total_bill", "tip").add_legend()

plt.show()
```

```python
# Now we will see multiple plots daywise and then genderwise
fg = sns.FacetGrid(df, col="day", hue="sex")


# Now plotting the dataset
fg.map(plt.scatter, "total_bill", "tip").add_legend()

plt.show()
```

```python
# Now we will see multiple plots daywise and then genderwise
fg = sns.FacetGrid(df, col="day", hue="sex")


# Now plotting the dataset using bar graph
fg.map(plt.bar, "total_bill", "tip").add_legend()

plt.show()
```
