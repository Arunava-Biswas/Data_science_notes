# HeatMap Plot

- To create Heatmap we use `heatmap()` with attribute `data` for the dataset.
- Heatmap mainly works on a 2 dimensional array. It will not work on 1 dimensional array.
- HeatMap is used to represent the correlation between each feature of the dataset using colours.
- Mainly used in ML algorithms to find Multicollinearity.
- Remember we cannot create Heatmap of string values, so to make Heatmap of categorical values first we need to convert their values into integers.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Creating basic Heatmap using numpy

```python
# Creating 20 elements and then create a 2 dimensional array.

var = np.linspace(1, 10, 20).reshape(4, 5)
var
```

```python
# now using heatmap

sns.heatmap(var)
```

### Now importing dataset

```python
dataset1 = sns.load_dataset("anagrams")
dataset1
```

```python
# dropping the features with non numeric values

x = dataset1.drop(columns=["attnr"], axis=1)
x
```

```python
# Creating a heatmap

sns.heatmap(x)
```

```python
# As the data points are too many so let's select only the 1st 10 data points

df = x.head(10)
df
```

```python
# Now let's create heatmap of this 10 data points

sns.heatmap(df)
```

**Parameter `vmin` and `vmax`**

- To set the minimum and maximum value of the heatmap.
- Like here we will set the minimum as 0 and maximum as 12.

```python
sns.heatmap(df, vmin=0, vmax=12)
```

**Parameter `cmap`**

- To change the colour of the heatmap.
- To find the possible combination do as early.

```python
sns.heatmap(df, vmin=0, vmax=12, cmap="PuOr")
```

**Parameter `annot`**

- To see the values of the heatmap.

```python
sns.heatmap(df, vmin=0, vmax=12, cmap="PuOr", annot=True)
```

**Parameter `annot` and `fmt`**

- To customize the values in the heatmap.
- First we need to create a list of values.
- Then we need to change the value of annot with our new customized array.
- Also we need to use the `fmt` parameter to specify the data type we are passing in the `annot` parameter.

```python
# Creating 2d dataframe

var1 = np.linspace(1, 10, 10).reshape(2, 5)
var1
```

```python
# now creating the heatmap

sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=True)
```

```python
# Now creating an array to customize the values

arr = np.array([["a0", "a1", "a2", "a3", "a4"], 
                ["b0", "b1", "b2", "b3", "b4"]])

arr
```

```python
# Now passing customized value in heatmap
# Here "fmt" is "s" as the datatype of array is string

sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s")
```

**Parameter `annot_kws`**

- To change multiple parameters like fontsize, color in one go.
- For this we need to create a dictionary first as we want to change multiple things at one go.

```python
# creating the dictionary
y = {"fontsize": 20, "color": "r"}

sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s", annot_kws=y)
```

**Parameter `linewidth`**

- To change the line width of the box, so we can seperate two blocks.
- This will create gaps between the boxes in the heatmap.

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s", annot_kws=y, linewidth=4)
```

**Parameter `linecolor`**

- To change the line colors instead of creating gaps between two boxes.

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s", annot_kws=y, linewidth=4, linecolor="g")
```

**Parameter `cbar`**

- To hide the color scale shown beside the heatmap.
- Give the value as `False` to hide the scale.

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s", annot_kws=y, 
            linewidth=4, linecolor="g", cbar=False)
```

**Parameter `xtickslabels` and `ytickslabels`**

- To hide all the labels.

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s", annot_kws=y, 
            linewidth=4, linecolor="g", cbar=False, xticklabels=False, yticklabels=False)
plt.show()
```

**Parameter `xtickslabels` and `ytickslabels`**

- To add labels in the heatmap.
- One way is to add `plt.xlabel()` and `plt.ylabel()`.
- But if we want to do through seaborn:
    - 1st we need to create a varibale as here it is `v` and pass the entire graph into it.
    - Then we need to use the `set()` along with that variable name.
    - Then pass the `xlabel` and `ylabel` along with their desired values we want to show.

```python
v = sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s", annot_kws=y, 
            linewidth=4, linecolor="g", cbar=False, xticklabels=False, yticklabels=False)
v.set(xlabel="python", ylabel="values")
plt.show()
```

**Parameter `font_scale`**

- To change the font color and fontsize of the labels.
- For fontsize we use `font_scale`

```python
v = sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=arr, fmt="s", annot_kws=y, 
            linewidth=4, linecolor="g", cbar=False, xticklabels=False, yticklabels=False)
v.set(xlabel="python", ylabel="values")
sns.set(font_scale=3)
plt.show()
```
