# Styling Plot

- This is to know what changes can we make during plotting so our plot becomes more presentable.
- Here we will know:
    - Seaborn Figure Styles
    - Removing Axes Spines
    - Scale and Context

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
df = sns.load_dataset("tips")
df
```

```python
# Now plotting using 'Boxplots'
# Here we will create single boxplot

sns.boxplot(y="total_bill", data=df)
plt.title("Vertical Box plot of Total Bill", fontsize=20)
plt.show()
```

```python
# To remove the name we use the 'plt.show()'

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill", fontsize=20)
plt.show()
```

#### Now to create grids

```python
# We can use 'plt.grid()' for this but it will create a basic grid only

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill", fontsize=20)
plt.grid()
plt.show()
```

#### Now creating grid using seaborn

**Parameter `set_style()`**

- Using this we can define the grid of our own choice.
- Always remember to set the grid before creating the plots.

```python
# Creating the grid
sns.set_style("darkgrid")

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill", fontsize=20)
plt.show()
```

#### Removing axes spine

**Parameter `despine`**

- This will remove the axes lines. That is it will remove the axis lines that we can see in this case at the right handside in the 1st graph.

```python
sns.set_style("whitegrid")

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill with axes spine", fontsize=20)

plt.show()
```

```python
sns.set_style("whitegrid")

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill without axes spine", fontsize=20)

sns.despine()
plt.show()
```

#### Scaling and context

**To change the figure size**

- We need to use the `figure()` of `matplotlib` library.
- Then to change the size we need to use the parameter `figsize`.

```python
sns.set_style("whitegrid")

# Here we want our figure with width 12 and height 3
plt.figure(figsize=(12, 3))

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill without axes spine", fontsize=20)

sns.despine()
plt.show()
```

#### Making changes to the labels

**Parameter `set_context()`**

- Here we can pass what kind of context we want in our graph.
- Here first we can create the style like here we pass the parameter `poster`.
- Then we can style the size of the fonts for this we will use `font_scale` parameter.
- To change colors we use `palette` parameters in the graph like we used in prtevious occassions.

```python
sns.set_style("whitegrid")

sns.set_context("poster", font_scale=2)

plt.figure(figsize=(12, 3))

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill without axes spine", fontsize=20)

sns.despine()
plt.show()
```

```python
# Here we using 'paper' style

sns.set_style("whitegrid")

sns.set_context("paper", font_scale=2)

plt.figure(figsize=(12, 3))

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Vertical Bar graph of Day vs Total Bill without axes spine", fontsize=20)

sns.despine()
plt.show()
```

```python
sns.set_style("whitegrid")

sns.set_context("paper", font_scale=2)

plt.figure(figsize=(12, 3))

sns.barplot(x="day", y="total_bill", data=df, palette="cool")
plt.title("Vertical Bar graph of Day vs Total Bill without axes spine", fontsize=20)

sns.despine()
plt.show()
```
