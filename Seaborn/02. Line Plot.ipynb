# Line Plot

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# creating datapoints

data1 = [1,2,3,4,5,6,7]
data2 = [2,4,5,3,1,7,6]

print("Printing the datapoints:\n")
print(f"First one is:\n {data1}")
print(f"\nSecond one is:\n {data2}")
```

**Now in matplotlib when we create a lineplot we used to pass the parameters `x` and `y` inside `pyplot()`**

```python
plt.plot(data1, data2)
plt.show()
```

**Now using the seaborn**

- Here to create a lineplot we need to use `lineplot()`
- There we need to pass 2 parameters `x` and `y` each having the datapoints between ehom we want to create the plot.
- Rember that Seaborn plot works on Dataset so if we run the code after the above part it will not show any plot.
- So we need to create a dataframe out of the data points to make the plot work.
- Then we need to pass the dataframe into the parameter `data`.

```python
# this will throw error

sns.lineplot(x="data1", y="data2")
```

**Creating the dataframes**

```python
df = pd.DataFrame({"data1": data1, "data2": data2})
df
```

**Now trying to do the plot again**

```python
sns.lineplot(x="data1", y="data2", data=df)
```

**To remove the label:**

- We need to use the matplotlib function `show()` to remove the label.

```python
sns.lineplot(x="data1", y="data2", data=df)
plt.show()
```

**The difference between the plot created by matplotlib and seaborn are:**

- The labels for `x-axis` and `y-axis` is predefined in seaborn unlike the matplotlib graph.

```python
# if we want to see the graph like in matplotlib then 
# pass the datapoints directly instead of using them inside ""

sns.lineplot(x=data1, y=data2)
```

### Now importing datasets

```python
dataset1 = sns.load_dataset("penguins")
dataset1
```

```python
# line graph between "bill_length_mm" and "bill_depth_mm"

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=dataset1)
plt.show()
```

```python
# line graph between "bill_length_mm" and "flipper_length_mm"

sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=dataset1)
plt.show()
```

**Parameter `hue`**

- This parameter is used when we want to create a graph in respect to the popular data in the dataset.

```python
# line graph between "bill_length_mm" and "bill_depth_mm" on the basis of sex

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset1, hue="sex")
plt.show()
```

**Notes:**

- Unlike matplotlib here we get the legend automatically.

**Parameter `size`**

- This parameter is used to customize the size.

```python
# line graph between "bill_length_mm" and "bill_depth_mm" 
# on the basis of sex with size of 10

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset1, hue="sex", size=10)
plt.show()
```

**Parameter `style`**

- This parameter is used to customize the style.
- Here we can change the style of each line.

```python
# line graph between "bill_length_mm" and "bill_depth_mm" 
# on the basis of sex with style of line showing female 
# as dotted line

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset1, hue="sex", style="sex")
plt.show()
```

**Parameter `patlette`**

- This parameter is used to customize the color.
- To see all the available palettes just enter any letter and press enter as a result it will throw an error but also will give all the available color combinations.

```python
# Checking the color combinations


sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset1, hue="sex", style="sex", 
             palette="A")
plt.show()
```

```python
# Here the color of the lines will change


sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset1, hue="sex", style="sex", 
             palette="Accent")
plt.show()
```

**Parameter `marker`**

- This parameter is used to get markers for each breaks on the line.
- In this parameter we can pass a list for multiple lines as in this case.

```python
# Here we will get dotted marker on Male for this we used "o" and arrow marker 
# for female for which we used ">"

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset1, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"])
plt.show()
```

**To see all the changes visibilly let's shorten down the dataset and then apply all the parameters of above**

```python
# here we use only the top 20 datapoints

dataset2 = sns.load_dataset("penguins").head(20)
dataset2
```

```python
# Now plotting

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset2, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"])
plt.show()
```

**Parameter `dashes`**

- To remove the dotted lines we will use this parameter with value as `False`.

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=dataset2, hue="sex", 
             style="sex", palette="Accent", markers=["o", ">"], dashes=False)
plt.show()
```

**Parameter `legend`**

- Here to remove the legends we need to use this parameter whereas in matplotlib to get the legends we used this parameter.
- To check the list of parameters in legend do same as in palette.

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset2, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], 
             dashes=False, legend=False)
plt.show()
```

**To make grid and give labels and title**

- To create grid use the `grid()` of matplotlib.
- To create title use the `title()` of matplotlib.
- For labels also use the `xlabel()` and `ylabel()` of matplotlib.

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", 
             data=dataset2, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False)

plt.grid()
plt.title("Graph between bill length and bill depth", fontsize=20)
plt.xlabel("bill length")
plt.ylabel("bill depth")

plt.tight_layout()
plt.show()
```
