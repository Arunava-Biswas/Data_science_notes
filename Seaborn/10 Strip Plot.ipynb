# Strip Plot

- It is basically a scatter plot that differentiates different categories.
- `Strip Plot` are a good alternative to a `Box Plot` or a `Violin Plot`.
- Here we use `stripplot()` along with parameters as `x`, `y`, and `data`.
- Here it shows all the distribution of plots differently in one go.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
df = sns.load_dataset("tips")
df
```

```python
# plotting the data

sns.stripplot(x="day", y="total_bill", data=df)
plt.title("Day vs Total Bill", fontsize=20)
plt.show()
```

**Parameter `hue`**

- To differentiate the graph.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex")
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```

**Parameter `palette`**

- To change the colors.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r")
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```

**Parameter `linewidth`**

- To change the data points thickness.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=2)
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```

**Parameter `edgecolor`**

- To change the edge colors of the `linewidth`.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=2, edgecolor="r")
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```

**Parameter `jitter`**

- To change the displacements between the datapoints.
- It creates space horizontally between datapoints.
- Remember the more value we put the more distributed the plot will be.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=2, edgecolor="r", jitter=3)
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```

**Parameter `size`**

- To change the size of the datapoints.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", 
              linewidth=2, edgecolor="r", jitter=1, size=10)
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```

**Parameter `marker`**

- To change the markers of the datapoints.
- We can also use a dictionary for different datapoints.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", 
              linewidth=2, edgecolor="r", jitter=1, size=10, marker="<")
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```

**Parameter `alpha`**

- To blur the graph.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", 
              linewidth=2, edgecolor="r", jitter=1, size=10, marker="<",
             alpha=0.4)
plt.title("Day vs Total Bill Genderwise", fontsize=20)
plt.show()
```



### To make stripplot of only one feature.

```python
sns.stripplot(x=df["total_bill"])
plt.title("Total Bill Stripplot", fontsize=20)
plt.show()
```

```python
# to make vertical

sns.stripplot(y=df["total_bill"])
plt.title("Total Bill Stripplot", fontsize=20)
plt.show()
```
