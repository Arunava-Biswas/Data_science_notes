# Bar Plot

- To create a bar plot we need to use `barplot()` along with two parameters `x` and `y` and the `data` attribute to refer the dataset.
- The bar plot always get created on the comaprison of two parameters `x` and `y`.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Now importing dataset

```python
dataset1 = sns.load_dataset("penguins")
dataset1
```

```python
# Now doing the plot using 'barplot()'
# here we are passing the x axis and y axis along with the dataframe name.

sns.barplot(x=dataset1.island, y=dataset1.bill_length_mm)
plt.title("Island vs bill_length", fontsize=20)
plt.show()
```

```python
# When passing the x axis and y axis directly then pass them as strings.
# Also pass the dataframe name as 'data' attribute.

sns.barplot(x='island', y='bill_length_mm', data=dataset1)
plt.title("Island vs bill_length", fontsize=20)
plt.show()
```

**Parameter `hue`**

- Now if we want to create more partition in the graph on the basis of some other feature we use this parameter.
- Like here we create a graph on the basis of gender for each island.
- So now we can differentiate how many `male` and `female` for each island.

```python
sns.barplot(x='island', y='bill_length_mm', data=dataset1, hue="sex")
plt.title("Island vs bill_length Gender wise", fontsize=20)
plt.show()
```

**Parameter `order`**

- This parameter is use to change the bar's positions.
- Like here we want the data to be shown in alphabetical order.

```python
# Checking the different names of the islands

dataset1["island"].unique()
```

```python
# Creating the order in which we want the data to be shown
order_1 = ["Biscoe", "Dream", "Torgersen"]

sns.barplot(x='island', y='bill_length_mm', 
            data=dataset1, hue="sex", order=order_1)
plt.title("Island vs bill_length Gender wise in alphabetical order", fontsize=20)
plt.show()
```

**Parameter `hue_order`**

- To change the order of the `hue` parameter.
- Like here if we want the female bar to come first.

```python
sns.barplot(x='island', y='bill_length_mm', 
            data=dataset1, hue="sex", order=order_1, 
            hue_order=["Female", "Male"])
plt.title("Island vs bill_length Gender wise in alphabetical order", fontsize=20)
plt.show()
```

**Parameter `ci`**

- Here the line we see on the top of the bar is for the Confidence Interval.
- To increase or decrease the values of the confidence interval we use this parameter. 
- The range varies between `0` to `100`.

```python
sns.barplot(x='island', y='bill_length_mm', 
            data=dataset1, hue="sex", order=order_1, 
            hue_order=["Female", "Male"], ci=100)
plt.title("Island vs bill_length Gender wise in alphabetical order", fontsize=20)
plt.show()
```

**Parameter `n_boot`**

- This parameter is used to show the confidence interval with gap.

```python
sns.barplot(x='island', y='bill_length_mm', 
            data=dataset1, hue="sex", order=order_1, 
            hue_order=["Female", "Male"], ci=100, n_boot=3)
plt.title("Island vs bill_length Gender wise in alphabetical order", fontsize=20)
plt.show()
```

### Creating horizontal bar graph


**Points to remember:**

- When creating a horizontal bar graph always remember that **both the `x` and `y` should have numerical values and not categorical values**.

**Parameter `orient`**

- To make the graph horizontal pass the value as `h` and for vertical we use `v`.

```python
sns.barplot(x='bill_depth_mm', y='bill_length_mm', data=dataset1, orient="h")
plt.title("bill_depth_mm vs bill_length horizotally", fontsize=20)
plt.show()
```

**Parameter `color`**

- To change the color of the bars.

```python
sns.barplot(x='island', y='bill_length_mm', data=dataset1, color='g')
plt.title("Island vs bill_length", fontsize=20)
plt.show()
```

**Parameter `palette`**

- To change the pattern of the colors of the bars.
- To know the available color combinations pass any letter and press enter and as result in the value error we will get all the available combinations.

```python
sns.barplot(x='island', y='bill_length_mm', 
            data=dataset1, hue="sex", palette="icefire")
plt.title("Island vs bill_length Gender wise", fontsize=20)
plt.show()
```

**Parameter `saturation`**

- To lighten (blur) the colors of the bar.

```python
sns.barplot(x='island', y='bill_length_mm', 
            data=dataset1, hue="sex", palette="icefire", 
            saturation=0.3)
plt.title("Island vs bill_length Gender wise", fontsize=20)
plt.show()
```

**Parameter `errcolor`**

- To change the colors of the errors i.e. the confidence intervals.

```python
# Changing the errors to yellow color

sns.barplot(x='island', y='bill_length_mm', 
            data=dataset1, hue="sex", palette="icefire", 
            saturation=0.3, errcolor='y')
plt.title("Island vs bill_length Gender wise", fontsize=20)
plt.show()
```

**Parameter `errwidth`**

- To change the width of the errors i.e. the confidence intervals.

```python
sns.barplot(x='island', y='bill_length_mm', data=dataset1, hue="sex", 
            palette="icefire", saturation=0.3, errcolor='y', errwidth=12)
plt.title("Island vs bill_length Gender wise", fontsize=20)
plt.show()
```

**Parameter `capsize`**

- To create a line to show the end point. Here using this we can create the capsize we want to see.
- It creates cap on both upside and downside. It mainly used in box plots in matplotlib.

```python
sns.barplot(x='island', y='bill_length_mm', data=dataset1, hue="sex", 
            palette="", errcolor='b', capsize=0.5)
plt.title("Island vs bill_length Gender wise", fontsize=20)
plt.show()
```

**Parameter `dodge`**

- To impose one graph upon another so we can compare them in a better way.
- Here we need to pass value as `False` to make the super impose happen.

```python
sns.barplot(x='island', y='bill_length_mm', data=dataset1, 
            hue="sex", order=order_1, hue_order=["Female", "Male"],
           dodge=False)
plt.title("Island vs bill_length Gender wise in alphabetical order", fontsize=20)
plt.show()
```

**Parameter `alpha`**

- To make the graph a bit blurry.

```python
sns.barplot(x='island', y='bill_length_mm', data=dataset1, 
            hue="sex", order=order_1, hue_order=["Female", "Male"],
           saturation=100, alpha=0.7)
plt.title("Island vs bill_length Gender wise in alphabetical order", fontsize=20)
plt.show()
```

### To add grid

- In case of seaborn we need to use the `set()` with parameter `style` and give value to that.
- Like here we want to use the `darkgrid` as value.

```python
sns.set(style="darkgrid")

sns.barplot(x='island', y='bill_length_mm', data=dataset1, 
            hue="sex", order=order_1, hue_order=["Female", "Male"],
           saturation=100, alpha=0.7)
plt.title("Island vs bill_length Gender wise in alphabetical order", fontsize=20)
plt.show()
```
