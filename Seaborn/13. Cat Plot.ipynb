# Cat Plot

- It is the advanced version of the `factorplot()`.
- Here we use the `catplot()` with `x`, `y`, `data` and `kind` parameters.
- Here we can create different plots using the `kind` parameter.


- **The plots that can be created using this plot are:**
   
   
   - **Categorical Scatter plots with `catplot`**
        - **`stripplot()`:** Here we need to pass the value **`strip`** into **`kind`** parameter.
        - **`swarmplot()`:** Here we need to pass the value **`swarm`** into **`kind`** parameter.
        
        
   - **Categorical Estimate plots with `catplot`**
       - **`pointplot()`:** Here we need to pass the value **`point`** into **`kind`** parameter.
       - **`barplot()`:** Here we need to pass the value **`bar`** into **`kind`** parameter.
       - **`countplot()`:** Here we need to pass the value **`count`** into **`kind`** parameter.
          
        
   - **Categorical Distribution plots with `catplot`**
       - **`boxplot()`:** Here we need to pass the value **`box`** into **`kind`** parameter.
       - **`violinplot()`:** Here we need to pass the value **`violin`** into **`kind`** parameter.
       - **`boxenplot()`:** Here we need to pass the value **`boxen`** into **`kind`** parameter.

```python
# Importing the libraries

import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Importing the dataset

```python
# taking only top 20 data points

df = sns.load_dataset("tips").head(20)
df
```

```python
# Now plotting

sns.catplot(x="tip", y="size", data=df)
plt.title("Normal Catplot of Tip vs Size", fontsize=20)
plt.show()
```

**Parameter `hue`**

- To differentiate the data points.

```python
sns.catplot(x="tip", y="size", data=df, hue="sex")
plt.title("Normal Catplot of Tip vs Size", fontsize=20)
plt.show()
```

**Parameter `palette`**

- To change the color of the data.

```python
sns.catplot(x="tip", y="size", data=df, hue="sex", palette="Oranges")
plt.title("Normal Catplot of Tip vs Size", fontsize=20)
plt.show()
```

**Parameter `height`**

- To change the height.

```python
sns.catplot(x="tip", y="size", data=df, hue="sex", palette="Oranges", height=3)
plt.title("Normal Catplot of Tip vs Size", fontsize=20)
plt.show()
```

### Now with the parameter `kind`

```python
sns.catplot(x="day", y="size", data=df, hue="sex", palette="Oranges", kind="bar")
plt.title("Bar graph of Day vs Size", fontsize=20)
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df, hue="sex", palette="Oranges", kind="point")
plt.title("Point plot of Day vs Size", fontsize=20)
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df, hue="sex", palette="Oranges", kind="box")
plt.title("Box plot of Day vs Size", fontsize=20)
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df, hue="sex", palette="Oranges", kind="boxen")
plt.title("Boxen plot of Day vs Size", fontsize=20)
plt.show()
```
