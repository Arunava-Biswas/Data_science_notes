## *Deep Learning*


- **Reasons behind the popularity of *Deep Learning*:**
  - **Growth in Data in recent years:** 
    - In recent years the volume of data generated by different businesses is huge compare to few years back. Also due to growth of social media the data generation grow exponencially. Now ***Deep Learning*** works better if the volume of data is higher.   
  - **Advancement in Hardware:**
    - With advanced hardwares in computers as specialized hardwares like ***GPU (Graphical Processing Unit)*** and ***TPU (Tensorflow Processing Unit)*** the works of ***Deep Learning*** becomes faster and easier now a days.    
  - **Python and Opensource Ecosystem:**
    - Now a days with the programming language like ***Python*** the ***Deep Learning*** programs can be written even without programming knowledge, it was not possible previously as earlier the programs are written in ***C++*** and ***JAVA*** which needed proper programming skills. Also ***Deep Learning*** Frameworks like ***Pytorch*** and ***Tensorflow*** makes the job easier.    
  - **Cloud and AI:**
    - Now a days we can run the program even in Clouds (renting a computer) without having a high end machine.

#### `Matrix Multiplication`

```python
import numpy as np
```

```python
# Creating two multi-dimensional arrays

revenue = np.array([[180,200,220], [24,36,40], [12,18,20]])
expenses = np.array([[80,90,100], [10,16,20], [8,10,10]])
```

```python
# Checking the matrixes

revenue
```

```python
expenses
```

```python
# Now calculating profits by simple subtraction of the two matrixes

profit = revenue - expenses
profit
```

```python
# Again creating matrixes for multiplication

price_per_unit = np.array([1000,400,1200])
units = np.array([[30,40,50], [5,10,15], [2,5,7]])
```

```python
price_per_unit
```

```python
units
```

```python
# now doing multiplication

price = price_per_unit*units
price
```

```python
# We can also do dot product using "dot()" of numpy

np.dot(price_per_unit, units)
```
