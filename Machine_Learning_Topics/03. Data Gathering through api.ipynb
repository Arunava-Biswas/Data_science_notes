### Here we will collect the data from an `api` of a website.

```python
# importing the libraries

import pandas as pd
import requests
```

```python
pd.reset_option('display.max_columns')
```

```python
# here we are collecting top rated movies data from the 'tmdb' website
# Creating http request from the url

response = requests.get('https://api.themoviedb.org/3/movie/top_rated?
                        api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
# Converting the data to json format

response.json()
```

```python
# Now we only need the 'results'

response.json()['results']
```

```python
# Now creating a dataframe of the result

pd.DataFrame(response.json()['results'])
```

```python
# Now we are selecting the columns we want along with the previous logics to create a dataframe

temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity',
                                                    'vote_average','vote_count']]
temp_df.head()
```

- Here we are getting the data from just the 1st page.
- Now if we want to get the data from all the pages we have to run the logic in a loop.
- So we need to create an empty dataframe.
- Then we need to append all the dataframes we get through the `requests` in the loop to this new dataframe.
- We can use `concat` or `merge` instead of `append`.

```python
# Creating an empty dataframe

df = pd.DataFrame()
df
```

```python
# Creating the loop to append data to this new dataframe
# here we use '.format(i)' we to get new page on each iteration
# 'ignore_index=True' so the index not get clashed with each other on each iteration

for i in range(1, 429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?
                            api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity',
                                                        'vote_average','vote_count']]
    df = df.append(temp_df, ignore_index=True)
```

```python
# Checking the dataset

df
```

```python
df.shape
```

```python
# Now saving the dataframe to a csv

try:
    df.to_csv('datasets/movies.csv', index=False)
except Exception as err:
    print("Error is: ", err)
else:
    print("Csv file created successfully.")
```
