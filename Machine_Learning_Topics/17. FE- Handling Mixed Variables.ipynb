### `Handling Mixed Variables`


- *`Mixed Data`* is a dataset where we have both *`Categorical`* and *`Numerical`* data in a column.
- As here in this example we can see with the column *`Cabin`* where there are two pieces of information attached with each other like *`B5`* means *`Class B`* that is *`Categorical`* data and *`Cabin no: 5`* that is *`Numerical`* data.
- So here first we have to split the value into two seperate columns one holding the *`Categorical`* information and other the *`Numerical`* information.
- Another way of having *`Mixed Data`* is a single column having data like [A, 5, G, 4], as here in example the column *`number`*. Here also we need to split the data in *`Categorical`* and *`Numerical`* columns.

```python
# Importing libraries

import pandas as pd
import numpy as np
```

```python
# importing the dataset

df = pd.read_csv('datasets/titanic.csv')
df.head()
```

```python
# Checking unique values in the column 'number' 
# Here the value 'A' represents Alone

df['number'].unique()
```

```python
# Creating bar plot to see how many times each category occured

fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers numbers with or without companion')
```

```python
# Creating a numeric column named 'number_numerical'
# Inside that column passing the original column 'number' using 'pd.to_numeric()'
# Also set the parameters: errors='coerce', downcast='integer'


df['number_numerical'] = pd.to_numeric(df["number"], errors='coerce', downcast='integer')
```

```python
# Creating a categorical column named 'number_categorical'
# Inside that column pass the numerical column we created before 'number_numerical' using the following logic:
# Wherever the numerical column has null value replace it with 'nan'

df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)

# Checking the final dataframe
df.head()
```

**Notes:**

- So now we can see that two extra columns get created by the column *`number`* on the basis of the value in the column.
- After doing this we can replace all the *`NaN`* values of the numerical column with *`0`*.

```python
# Applying same logic in 'Cabin' and 'Ticket' columns 


df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

**Notes**

- So here we can see too many categories to handle so we apply the logic of differentiating the alphabets from the numbers and create two separate columns for them.

```python
# captures numerical part
# str.extract('(\d+)') to extract the numeric part of the data
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') 

# captures the first letter
df['cabin_cat'] = df['Cabin'].str[0] 

df.head()
```

```python
# Now we can see reduced categories in cabin

df['cabin_cat'].value_counts()
```

```python
# Checking using bar plot we can see most people are in cabin C

df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# Applying same logic for the column 'Ticket'

# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```

```python
# Now checking number of categories in ticket

df['ticket_cat'].unique()
```
