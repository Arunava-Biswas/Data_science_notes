- **Feature Engineering** is the process of using domain knowledge to extract features from raw data. These features can be used to improve the performance of ML algorithms.
- It is a very important topic.
- Here there is no fixed rule.




- Here there are 4 main topics in **FE**.

    A. **Feature Transformation**: Here we transform a feature (column) from one form to another form so it can perform better when training the model. Here we follow some more steps:
     1. **Missing Value Imputation**: Here we replace the missing values in the dataset as **`Scikit-Learn`** doesn't accept missing values when training the model. So we need to either replace or delete the missing values before training the model.
     2. **Handling Categorical Features**: Here we need to transform the categorical data to numeric data or numeric data to categorical data **(Binning)**. For this we use **`One Hot Encoding`** or **`Label Encoding`** or **`Ordinal Encoding`**.
     3. **Outlier Detection**: To detect the outliers in the dataset and remove or replace them from the dataset.
     4. **Feature Scaling**: To **Standardize** and **Normalize** the values of different features so all the features are calculated under same range. It helps in finding ucledian distance between two points. So unless both the points of same scale there will be an error in the model.
        
        
   B. **Feature Construction**: Here we manually create some features (columns) which are not present in the dataset by default. It is done so the algorithm can perform in a more better way. Here we create a complete new column manually like we want the hours `hh` out of a column holing time in `hh:mm` format. Here there are different techniques like **Feature Splitting**, **Grouping** etc.
    
    
   C. **Feature Selection**: Here we select which features (columns) to pass to the algorithm out of all the features of the dataset. That is here we select only those features which are very important for the performance of the model. This solves the problem of **Curse of Dimensionality**. Here also some techniques involved like **Backward Selection**, **Forward Elimination** etc.
    
    
   D. **Feature Extraction**: Here we programatically and not manually extract completely new features out of the given features in the dataset. This is different from **Feature Construction**. Here also we have some techniques like **PCA** etc.
